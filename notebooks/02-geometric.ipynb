```python
import matplotlib.pyplot as plt
import numpy as np

# --- 1. Define the Data (Tesla Model Y Depreciation) ---
# n represents Years 1 through 10 (Project requires 5-10 terms)
n = np.arange(1, 11)

# Formula: a_n = a * r^(n-1)
# a = 40,000 (Start Price)
# r = 0.8 (Retaining 80% value)
# Note: In Python, exponents use '**'
car_value = 40000 * (0.8) ** (n - 1)

# --- 2. Create the Plot ---
plt.figure(figsize=(8, 6))

# Plot the points with a line connecting them
# 'marker' adds dots at the specific years calculated
# 'color' distinguishes this from your blue Arithmetic graph
plt.plot(n, car_value, marker='o', linestyle='-', color='red', linewidth=2, label='Car Value (20% Depr.)')

# --- 3. Style the Graph ---
plt.title('Geometric Sequence: Tesla Depreciation', fontsize=14)
plt.xlabel('Year (n)', fontsize=12)
plt.ylabel('Value ($)', fontsize=12)

# Add a grid to make reading values easier
plt.grid(True, linestyle='--', alpha=0.6)

# Force x-axis to show integer years only (1, 2, 3...)
plt.xticks(np.arange(1, 11, 1))

# --- 4. Annotations (Optional but helpful for the report) ---
# Highlight the starting price
plt.annotate('Purchase Price: $40k', 
             xy=(1, 40000), 
             xytext=(2.5, 38000),
             arrowprops=dict(facecolor='black', shrink=0.05))

# Highlight the "Half Life" roughly (around year 4)
plt.annotate('~50% Value', 
             xy=(4, 20480), 
             xytext=(5, 25000),
             arrowprops=dict(facecolor='black', shrink=0.05))

# --- 5. Save and Show ---
plt.legend()
# Save the image to your images folder to include in the Markdown report
plt.savefig('../images/gp_graph.png', dpi=300) 
plt.show()
```
