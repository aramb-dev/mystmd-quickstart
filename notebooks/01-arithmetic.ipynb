```python
import matplotlib.pyplot as plt
import numpy as np
# n represents Weeks 1 through 20
n = np.arange(1, 21) 

# Formula: a_n = 100 + (n-1)10
savings = 100 + (n - 1) * 10
plt.figure(figsize=(8, 5))

# Plot the points (scatter) and the line
plt.plot(n, savings, marker='o', linestyle='-', color='blue', label='Savings Balance')
plt.title('Arithmetic Sequence: Saving for MacBook', fontsize=14)
plt.xlabel('Week Number (n)', fontsize=12)
plt.ylabel('Total Savings ($)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(np.arange(1, 21, 2))  # Show every 2nd week on x-axis
# Point to the start
plt.annotate('Start: $100', xy=(1, 100), xytext=(3, 120),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.legend()
plt.savefig('../images/ap_graph.png') # Saves to your images folder
plt.show()
```
